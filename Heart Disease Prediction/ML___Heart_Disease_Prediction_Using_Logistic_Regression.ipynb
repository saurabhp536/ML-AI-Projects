```python
import pandas as pd
import pylab as pl
import numpy as np
import scipy.optimize as opt
import statsmodels.api as sm
from sklearn import preprocessing
'exec(% matplotlib inline)'
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import seaborn as sns
```

```python
# dataset
disease_df = pd.read_csv("framingham.csv")
disease_df.drop(['education'], inplace=True, axis=1)
disease_df.rename(columns={'male': 'Sex_male'}, inplace=True)
```

```python
# removing NaN / NULL values
disease_df.dropna(axis=0, inplace=True)
print(disease_df.head(), disease_df.shape)
print(disease_df.TenYearCHD.value_counts())
```

```python
from sklearn.model_selection import train_test_split
X = np.asarray(disease_df[['age', 'Sex_male', 'cigsPerDay',
                           'totChol', 'sysBP', 'glucose']])
y = np.asarray(disease_df['TenYearCHD'])

# normalization of the dataset
X = preprocessing.StandardScaler().fit(X).transform(X)

# Train-and-Test -Split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=4)

print('Train set:', X_train.shape,  y_train.shape)
print('Test set:', X_test.shape,  y_test.shape)
```

```python
# counting no. of patients affected with CHD
plt.figure(figsize=(7, 5))
sns.countplot(x='TenYearCHD', data=disease_df,
              palette='coolwarm')
plt.show()
```

```python
laste = disease_df['TenYearCHD'].plot()
plt.show(laste)
```

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
y_pred = logreg.predict(X_test)
```

```python
# Evaluation and accuracy
from sklearn.metrics import accuracy_score
print('Accuracy of the model is =',
      accuracy_score(y_test, y_pred))
```

```python
# Confusion matrix
from sklearn.metrics import confusion_matrix, classification_report

cm = confusion_matrix(y_test, y_pred)
conf_matrix = pd.DataFrame(data=cm,
                           columns=['Predicted:0', 'Predicted:1'],
                           index=['Actual:0', 'Actual:1'])

# Bar plot visualization  
conf_matrix.plot(kind='bar', figsize=(8, 5), color=['Green', 'Red'], alpha=0.7)  
plt.title('Confusion Matrix as Bar Plot')  
plt.ylabel('Count')  
plt.xlabel('Actual Classes')  
plt.xticks(rotation=0)  # Keeps the labels readable  
plt.legend(title='Predicted Classes')  
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Adds horizontal grid lines for clarity  

# Show the plot  
plt.show()  

# Print classification report  
print('The details for confusion matrix is =')  
print(classification_report(y_test, y_pred))
```
